```python
import json
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.otomoto.pl/osobowe"
pages = [f"{url}?page={x}" for x in range(0,6250)]
```

```python
def get_params(bs_tag, link):
    try:
        price = bs_tag.find("span", {"class":"offer-price__number"}).text.strip().replace(" ", "")
    except AttributeError:
        price = "lack of price"
    keys = ["link", "price"]
    values = [link["href"], price]
    params = bs_tag.find("div", {"class":"parametersArea"})
    try:
        for par in params.find_all("li"):
            try:
                keys.append(par.span.text)
                values.append(par.div.text.lstrip().rstrip())
                # print("\n\n\n")
            except AttributeError:
                pass
            zipped = dict(zip(keys, values))
    except AttributeError:
        zipped = {}
    return zipped
```

```python
for page in pages:
    body = bs(requests.get(page).content).body
    for link in body.find_all("a", {"href":True, "class":None, "rel":"noreferrer", "target":"_self"}):
        small_body = bs(requests.get(link["href"]).content).body
        params = get_params(small_body, link)
        with open("jsonfile.txt", "a", encoding="utf-8") as f:
            f.write(f"{params};")
```
