```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
otomoto_df = pd.read_csv("dataset.csv")
otomoto_df
```

```python
grouped_brands = otomoto_df.groupby('marka_pojazdu')['marka_pojazdu'].count()
top10_brands = grouped_brands.sort_values(ascending=False)[:10].to_dict()
not_top10_brands = grouped_brands.sort_values(ascending=True)[:10].to_dict()
```

### Zobaczmy jak prezentuje się TOP 10 Najpopularnijeszych marek samochodowych w serwisie otomoto

```python
plot = sns.barplot(x=list(top10_brands.keys()), y=list(top10_brands.values()) )
plt.setp(plot.get_xticklabels(), rotation=40, horizontalalignment='center')
plt.show()
```

### A jak prezentuje się TOP 10 Najmniej popularnych 

```python
plot = sns.barplot(x=list(not_top10_brands.keys()), y=list(not_top10_brands.values()) )
plt.setp(plot.get_xticklabels(), rotation=40, horizontalalignment='center')
plt.show()
```

### Możemy zbadać sprawę samochodów zabytkowych...

```python
grouped_antiques = otomoto_df.groupby('czy_zabytek')['czy_zabytek'].count()
grouped_antiques
```

```python
antiques_sub_df = otomoto_df[(otomoto_df['czy_zabytek']=='Tak') & (otomoto_df['rok_produkcji']<1997)]
antiques_sub_df
```

```python
grouped_brands = antiques_sub_df.groupby(['marka_pojazdu', 'model_pojazdu', 'rok_produkcji', 'cena']).count()
grouped_brands.sort_values('cena', ascending=False)
```
