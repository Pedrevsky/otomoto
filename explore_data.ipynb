```python
to_df = list(set(all_keys))
df = pd.DataFrame(columns=to_df, data=dicts)
df = df.drop_duplicates()
```

```python
def wyjasnij_tomka(x):
    if x in ('Tak','Polska', 'Francja', 'Tomasz Barnaś') or pd.isna(x) or x=='' or x is None: return 0
    try:
        x = str(x).split( )[0]
    except: return x
    x = x.replace('l', '').replace('L', '').replace(',', '.').replace('/', '-')
    try:        
        x = float(x.split( )[0])
    except:
        # print(x)
        if len(x)>2:
            digits = x.split('-')
            if len(digits[0])>1:
                return float(digits[0])/10
            digits = [float(x) for x in digits]
            avg_x = sum(digits)/len(digits)
            x= avg_x
        
    
    return x/10 if x>20 else x
```
