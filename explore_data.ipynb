```python
import json
import pandas as pd
```

```python
with open("jsonfile.txt", "r", encoding="utf-8", errors='ignore') as f:
    content = f.read()

content = [el.replace("\'", "\"") for el in content.split(";")]
```

```python
dicts = []
with_error = []
for el in content:
    try:
        dicts.append(json.loads(el))
    except:
        with_error.append(el)
        # json.loads(f"r{el}")
```

```python
all_keys = []

for d in dicts:
    all_keys += list(d.keys())
```

```python
to_df = list(set(all_keys))
df = pd.DataFrame(columns=to_df, data=dicts)
df = df.drop_duplicates()
```
